```python
# Create an LLM-based agent
llm = ChatOpenAI(temperature=0, model="gpt-4o")  # Change model if needed
```

```python
def get_llm_response(question):
    return llm.invoke(question).content
```

```python
drugDiseaseDMDB = pd.read_csv('./data/DMDB_go_bp_filtered_jjoy_05_08_2025.csv')
```

```python
drugDiseaseSet = drugDiseaseDMDB.sample(50)

print(drugDiseaseSet)
```

```python
drugDiseaseSet['question'] = "Which drugs can treat " + drugDiseaseSet['disease_name'] + " by targeting " + drugDiseaseSet['bp_name'] + "? Return a comprehensive list that ONLY contains drugs that target the correct biological process"
```

```python
drugDiseaseSet['gpt4.0_response'] = drugDiseaseSet["question"].apply(get_llm_response)
```
